```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()
```

```python
#clf = svm.SVR(kernel='linear', epsilon=0.2, C=2)
#clf = svm.SVR(kernel='poly', degree=3, epsilon=0.2, C=2)
clf = svm.SVR(kernel='rbf', gamma=0.5, epsilon=0.2, C=2)
y_pred = clf.fit(X, y).predict(X)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='b')
plt.show()
```
