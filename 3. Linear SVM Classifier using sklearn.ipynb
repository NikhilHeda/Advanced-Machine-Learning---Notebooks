```python
from sklearn import svm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('svm_data.csv', names=['x1', 'x2', 'y'])
train_x = data[['x1','x2']]
train_y = data['y']
```

```python
# Fitting svm
clf = svm.SVC(kernel='linear')

clf.fit(train_x, train_y)

W = clf.coef_[0]
b = clf.intercept_[0]

xx = np.linspace(0, 12)
yy = -b/W[1] - xx*W[0]/W[1]
```

```python
data_pos = data[data.y == 1]
data_neg = data[data.y == -1]

plt.plot(data_pos.x1, data_pos.x2, 'b+', label='Positive Samples')
plt.plot(data_neg.x1, data_neg.x2, 'rx', label='Negative Samples')
plt.plot(xx, yy, label='Decision Boundary')
plt.title('Support Vector Machine - Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```
