```python
import tensorflow as tf
```

```python
message = tf.constant('Hello, World')

session = tf.Session()
ret_val = session.run(message)
#print(ret_val) # returns a bytearray
print(ret_val.decode())
session.close()
```

```python
a = tf.constant(4.0)
b = tf.constant(5.2)
add_op = tf.add(a, b) # (or) a + b

session = tf.Session()
ret_val = session.run(add_op)
print(ret_val)
session.close()
```

```python
a = tf.constant(3)
b = tf.constant(4)
c = tf.constant(5)

add_op = a + b # tf.add(a, b)
mul_op = add_op * c # tf.multiply(add_op, b)
pow_op = mul_op ** 2 # tf.pow(mul_op, 2)

session = tf.Session()
'''
print(session.run(add_op))
print(session.run(mul_op))
print(session.run(pow_op))
'''

print(session.run([add_op, mul_op, pow_op]))
```

```python
message = tf.constant('Hello, World')

with tf.Session() as sess:
    print(message.eval())
```

```python
message = tf.constant('Hello, World')

isess = tf.InteractiveSession()
print(message.eval())
isess.close()
```

```python
node1 = tf.constant([3, 4, 3, 4], shape=(2, 2))
node2 = tf.constant([4, 5, 4, 5], shape=(2, 2))
node3 = tf.matmul(node1, node2)

with tf.Session() as sess:
    print(sess.run(node3))
```

```python
num1 = tf.placeholder(dtype=tf.int32)
num2 = tf.placeholder(dtype=tf.int32)

add_op = num1 + num2

with tf.Session() as sess:
    result = add_op.eval(feed_dict={num1: 10, num2: 11})
    print(result)
```

```python
num1 = tf.Variable(5, name='a')
num2 = tf.Variable(10, name='b')

init_op = tf.global_variables_initializer()

assign_num1_op = tf.assign(num1, 50)
assign_num2_op = tf.assign(num2, 100)

add_op = tf.add(assign_num1_op, assign_num2_op)

with tf.Session() as sess:
    sess.run(init_op)
    
    print('Initial Values:', sess.run([num1, num2]))
    
    print('Reassigned values and sum:', sess.run([assign_num1_op, assign_num2_op, add_op]))
```

```python
sum_op = tf.reduce_sum([[2, 3, 4], [3, 4, 5]], axis=1)

with tf.Session() as sess:
    print(sess.run(sum_op))
```

```python
x = tf.Variable(2.0, dtype=tf.float32) # 0 does not work because internal update multiplies with x so weights never update
learning_rate = 0.001
momentum = 0.01

loss_f = tf.pow(x, 4) - (3 * tf.pow(x, 2)) - 2

#optimizer = tf.train.GradientDescentOptimizer(learning_rate)
optimizer = tf.train.MomentumOptimizer(learning_rate, momentum)
loss_op = optimizer.minimize(loss_f)

init_op = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    
    for _ in range(1000): # Until convergence, fixed epochs, acceptable loss, no change (stuck at local minima - diff from convergence)
        print(sess.run([x, loss_f]))
        sess.run(loss_op)
```
